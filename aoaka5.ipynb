```python
wb = load_workbook(savename)
ws = wb.worksheets[0]

maxrow = ws.max_row +1

#最大行調べ
col_test = 2

#行を逆ループ
for i in reversed(range(1,maxrow)):
    #セルがNoneだったら
    if ws.cell(row=i, column=col_test).value != None:
        maxrow = i
        break

maxcol = ws.max_column

#1列ずつ結合していく
#　A列
A_ketsugou_cell_e = []
for i in ws['A2:' + 'A' + str(maxrow)] :
    for j in i :
        if j.value != j.offset(1,0).value :
            A_ketsugou2 = j.coordinate
            A_ketsugou_cell_e.append(A_ketsugou2)
A_ketsugou_cell_s = []
for i in reversed(ws['A2:' + 'A' + str(maxrow)]) :
    for j in i :
        if j.value != j.offset(-1,0).value :
            A_ketsugou2 = j.coordinate
            A_ketsugou_cell_s.append(A_ketsugou2)
A_ketsugou_cell_s = [i for i in reversed(A_ketsugou_cell_s)]

# A列結合する
for i,j in zip(A_ketsugou_cell_s,A_ketsugou_cell_e):
    ws.merge_cells(i + ':' + j)

#　B列
B_ketsugou_cell_e = []
for i in ws['B2:' + 'B' + str(maxrow)] :
    for j in i :
        if j.value != j.offset(1,0).value :
            B_ketsugou2 = j.coordinate
            B_ketsugou_cell_e.append(B_ketsugou2)
B_ketsugou_cell_s = []
for i in reversed(ws['B2:' + 'B' + str(maxrow)]) :
    for j in i :
        if j.value != j.offset(-1,0).value :
            B_ketsugou2 = j.coordinate
            B_ketsugou_cell_s.append(B_ketsugou2)
B_ketsugou_cell_s = [i for i in reversed(B_ketsugou_cell_s)]

# B列結合する
for i,j in zip(B_ketsugou_cell_s,B_ketsugou_cell_e):
    ws.merge_cells(i + ':' + j)

#行を結合する
ws.merge_cells(range_string='D1:F1')
ws.merge_cells(range_string='G1:I1')
ws.merge_cells(range_string='J1:L1')
ws.merge_cells(range_string='M1:O1')
ws.merge_cells(range_string='P1:R1')
ws.merge_cells(range_string='S1:U1')
ws.merge_cells(range_string='V1:X1')
ws.merge_cells(range_string='Y1:AA1')
ws.merge_cells(range_string='AB1:AD1')
ws.merge_cells(range_string='AE1:AG1')
ws.merge_cells(range_string='AH1:AJ1')
ws.merge_cells(range_string='AK1:AM1')
ws.merge_cells(range_string='AN1:AP1')
ws.merge_cells(range_string='AQ1:AS1')
ws.merge_cells(range_string='AT1:AV1')
ws.merge_cells(range_string='AW1:AY1')
ws.merge_cells(range_string='AZ1:BB1')
ws.merge_cells(range_string='BC1:BE1')
ws.merge_cells(range_string='BF1:BH1')
ws.merge_cells(range_string='BI1:BK1')
ws.merge_cells(range_string='BL1:BN1')

#値の削除
for row in range(2,maxrow+1):
    for col in range(4,maxcol+1):
        ws.cell(row=row,column=col).value = np.nan

#　罫線を引く
side = Side(style='thin', color='808080')
border = Border(top=side, bottom=side, left=side, right=side)

for row in range(1,maxrow+1):
    for col in range(1,maxcol+1):
        ws.cell(row=row,column=col).border = border

#ウィンドウ枠の固定
ws.freeze_panes = 'D2'

ws.auto_filter.ref = get_column_letter(1) + str(1) + ':' \
    + get_column_letter(3) + str(1)


#セルに色付け
c_main = '5b9bd5'
c_0ji = '4472c4'
c_1ji = '70ad47'
c_2ji = 'ed7d31'
c_3ji = 'ffb700'
c_kashika = 'aa72d4'
c_full = 'bf8f00'
c_yotei = 'bdd7ee'
c_ref = '000000'

c_koutei = '50A8FF'
c_syutsu = 'CCCCFF'
c_chukan = 'EAEAFF'
c_compo = 'EAFFFF'
c_taikyu = 'FFEAEA'
c_hokan = 'EACCFF'
c_jisseki = 'D3DCE9'

side_main = Side(style='thin',color= c_main)
side_0ji = Side(style='thin',color= c_0ji)
side_1ji = Side(style='thin',color= c_1ji)
side_2ji = Side(style='thin',color= c_2ji)
side_3ji =  Side(style='thin',color= c_3ji)
side_ref = Side(style='thin',color=c_ref)
side_kashika = Side(style='thin',color=c_kashika)

fill_main = PatternFill(patternType='solid',fgColor= c_main)
fill_0ji = PatternFill(patternType='solid',fgColor= c_0ji)
fill_1ji = PatternFill(patternType='solid',fgColor= c_1ji)
fill_2ji = PatternFill(patternType='solid',fgColor= c_2ji)
fill_3ji = PatternFill(patternType='solid',fgColor= c_3ji)
fill_cae = PatternFill(patternType='solid',fgColor= c_1ji)
fill_kashika = PatternFill(patternType='solid',fgColor= c_kashika)
fill_full = PatternFill(patternType='solid',fgColor= c_full)
fill_db = PatternFill(patternType='solid',fgColor= c_3ji)
fill_new = PatternFill(patternType='solid',fgColor= c_main)
fill_yotei =  PatternFill(patternType='solid',fgColor= c_yotei)

fill_koutei =  PatternFill(patternType='solid',fgColor= c_koutei)
fill_syutsu =  PatternFill(patternType='solid',fgColor= c_syutsu)
fill_chukan =  PatternFill(patternType='solid',fgColor= c_chukan)
fill_compo =  PatternFill(patternType='solid',fgColor= c_compo)
fill_taikyu =  PatternFill(patternType='solid',fgColor= c_taikyu)
fill_hokan =  PatternFill(patternType='solid',fgColor= c_hokan)
fill_jisseki =  PatternFill(patternType='solid',fgColor= c_jisseki)

```
